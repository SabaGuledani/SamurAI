```python
import os
import pandas as pd
```

```python
directory_path = './datasets/preprocessed/'
filenames = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
```

```python
df = pd.read_csv(directory_path+filenames[0])
```

```python
columns = df.columns
```

```python
for i,file in enumerate(filenames):
    if i == 0:
        continue
    df_import = pd.read_csv(directory_path+filenames[i])
    df = pd.concat([df,df_import],ignore_index=True)
```
