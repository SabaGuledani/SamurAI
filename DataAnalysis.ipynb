# Analysis of processed data
interesting analysis of statistics on how certain extracted features corelate with link being benign or not

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('final_dataset.csv')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
len(df.dtypes)
```

```python
benign = df['benign'].values
domain_len = df['domain_len'].values
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = df.drop(columns=['url'])
# Calculate the correlation matrix

correlations = df.corr()

# Extract the correlation of each feature with the 'benign' column
correlation_with_benign = correlations['benign'].sort_values(ascending=False)
# print(correlation_with_benign)

# Plot the correlations
plt.figure(figsize=(10, 6))
correlation_with_benign.drop('benign').plot(kind='bar')
plt.title('Correlation of Features with Target Variable (Benign or Not)')
plt.xlabel('Features')
plt.ylabel('Correlation with Target (Benign)')
plt.show()


```
